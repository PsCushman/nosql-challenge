### 1. Which establishments have a hygiene score equal to 20?

```python
query = {"scores.Hygiene": 20}
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
num_documents = establishments.count_documents(query)
print(f"Number of establishments with hygiene score 20: {num_documents}")

# Display the first document in the results using pprint
if num_documents > 0:
    first_document = results[0]
    pprint(first_document)
else:
    print("No establishments found with hygiene score 20.")
```

```python
query = {"scores.Hygiene": 20}
results = establishments.find(query)

# Convert the result to a Pandas DataFrame
safe_resto_df = pd.DataFrame(list(results))

# Display the number of rows in the DataFrame
num_rows = len(safe_resto_df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
safe_resto_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    'LocalAuthorityName': {'$regex': 'London'},
    "RatingValue": {"$gte": 4}
}

results = establishments.find(query)

# Use count_documents to display the number of documents in the result
num_documents = establishments.count_documents(query)
print(f"Number of documents: {num_documents}")

# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
results = establishments.find(query) #talk to kourt

# Convert the result to a Pandas DataFrame
high_rated_london = pd.DataFrame(list(results))

# Display the number of rows in the DataFrame
num_rows = len(high_rated_london)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
high_rated_london.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

restaurant_name = "Penang Flavours"

# Find the restaurant document
restaurant_query = {
    "BusinessName": restaurant_name
}
restaurant_document = establishments.find_one(restaurant_query)


# Extract the latitude and longitude values
degree_search = 0.01
latitude = restaurant_document["geocode"]["latitude"] #talk to kourt
longitude = restaurant_document["geocode"]["longitude"]

query = {
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
    "RatingValue": 5
}

sort = [("scores.Hygiene", 1)]

results = establishments.find(query).sort(sort).limit(5)

# Print the results
for result in results:
    pprint(result)
```

```python
results = establishments.find(query).sort(sort)

# Convert result to Pandas DataFrame
penang_neighbors = pd.DataFrame(results)

# Display the DataFrame
penang_neighbors
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Execute the aggregation pipeline and store the result in a cursor
result = establishments.aggregate(pipeline)

# Iterate over the cursor and print each document
# counter = 0
# for document in result:
#     pprint(document)
#     counter += 1
#     if counter == 10:
#         break

#use itertools to slice at 10 results        #talk to kourt
for document in itertools.islice(result, 10):
    pprint(document)
```

```python
# Execute the aggregation pipeline and store the result in a cursor
result = establishments.aggregate(pipeline)

# Convert the result to a DataFrame
gross_rest_locs = pd.DataFrame(result)

# Display the first 10 rows of the DataFrame
gross_rest_locs.head(10)
```
